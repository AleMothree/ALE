```
# src/modules/database/database.py
import sqlite3
from datetime import datetime

class Database:
    """Handles database operations for logging and trade tracking."""

    def __init__(self, db_name='alemot.db'):
        """Initializes the database connection and creates tables if they don't exist."""
        self.conn = sqlite3.connect(db_name)
        self.cursor = self.conn.cursor()
        self._create_tables()

    def _create_tables(self):
        """Creates the necessary tables for logs and trades."""
        self.cursor.execute('''
            CREATE TABLE IF NOT EXISTS logs (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                timestamp TEXT NOT NULL,
                level TEXT NOT NULL,
                message TEXT NOT NULL
            )
        ''')
        self.cursor.execute('''
            CREATE TABLE IF NOT EXISTS trades (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                timestamp TEXT NOT NULL,
                exchange TEXT NOT NULL,
                symbol TEXT NOT NULL,
                type TEXT NOT NULL,
                side TEXT NOT NULL,
                amount REAL NOT NULL,
                price REAL,
                order_id TEXT,
                status TEXT
            )
        ''')
        self.conn.commit()

    def log(self, level, message):
        """Records a log message in the database."""
        timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        self.cursor.execute("INSERT INTO logs (timestamp, level, message) VALUES (?, ?, ?)",
                            (timestamp, level, message))
        self.conn.commit()

    def log_trade(self, trade_data):
        """
        Records a trade in the database.
        trade_data should be a dictionary with keys like:
        'exchange', 'symbol', 'type', 'side', 'amount', 'price', 'id' (order_id), 'status'
        """
        timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        self.cursor.execute(
            "INSERT INTO trades (timestamp, exchange, symbol, type, side, amount, price, order_id, status) "
            "VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)",
            (timestamp,
             trade_data.get('exchange', 'unknown'),
             trade_data.get('symbol', 'unknown'),
             trade_data.get('type', 'unknown'),
             trade_data.get('side', 'unknown'),
             trade_data.get('amount', 0.0),
             trade_data.get('price', None),
             trade_data.get('id', None),
             trade_data.get('status', 'unknown'))
        )
        self.conn.commit()

    def close(self):
        """Closes the database connection."""
        self.conn.close()

if __name__ == '__main__':
    # Example usage:
    db = Database()
    db.log('INFO', 'Database initialized successfully.')
    db.log('WARNING', 'This is a test warning.')

    # Simulate a trade
    sample_trade = {
        'exchange': 'binance',
        'symbol': 'BTC/USDT',
        'type': 'market',
        'side': 'buy',
        'amount': 0.001,
        'price': 65000.0,
        'id': 'test_order_123',
        'status': 'filled'
    }
    db.log_trade(sample_trade)
    db.log('INFO', 'Sample trade logged.')

    db.close()
    print("Database connection closed.")
```

```
import pandas as pd
import numpy as np
import time
import random

# --- 1. Data Acquisition Component (Basic Simulation) ---
class DataAcquisition:
    """Simulates fetching historical market data."""
    def __init__(self, symbol='BTC/USDT', start_date='2023-01-01', end_date='2023-12-31'):
        self.symbol = symbol
        self.start_date = start_date
        self.end_date = end_date
        self.data = self._generate_simulated_data()

    def _generate_simulated_data(self):
        """Generates simple simulated historical price data."""
        # Use 'h' instead of 'H' for hourly frequency as suggested by the warning
        date_rng = pd.date_range(start=self.start_date, end=self.end_date, freq='h')
        n_points = len(date_rng)
        # Simulate a price trend with some noise
        prices = np.linspace(30000, 70000, n_points) + np.random.randn(n_points) * 500
        df = pd.DataFrame(date_rng, columns=['timestamp'])
        df['price'] = prices
        df['volume'] = np.random.rand(n_points) * 100 # Simulated volume
        df.set_index('timestamp', inplace=True)
        return df

    def get_historical_data(self):
        """Returns the simulated historical data."""
        print(f"Fetching simulated historical data for {self.symbol}...")
        return self.data

    def get_latest_price(self):
        """Simulates getting the latest price."""
        if not self.data.empty:
            # Get the last price safely
            latest_price = self.data['price'].iloc[-1]
            print(f"Simulating latest price fetch: {latest_price:.2f}")
            return latest_price
        return None

# --- 2. Strategy Execution Component (Simple Moving Average Crossover) ---
class StrategyExecution:
    """Implements a simple trading strategy (SMA Crossover)."""
    def __init__(self, short_window=50, long_window=200):
        self.short_window = short_window
        self.long_window = long_window
        self.signals = pd.DataFrame()

    def generate_signals(self, data):
        """Generates buy/sell signals based on SMA crossover."""
        if data.empty or len(data) < max(self.short_window, self.long_window):
            print("Not enough data to generate signals.")
            self.signals = pd.DataFrame(index=data.index)
            self.signals['signal'] = 0
            return self.signals[['signal']] # Return signal column even if empty

        print("Generating trading signals...")
        signals = pd.DataFrame(index=data.index)
        signals['signal'] = 0.0

        # Calculate moving averages
        signals['short_mavg'] = data['price'].rolling(window=self.short_window, min_periods=1).mean()
        signals['long_mavg'] = data['price'].rolling(window=self.long_window, min_periods=1).mean()

        # Create signals using .iloc for positional slicing
        signals.iloc[self.long_window:, signals.columns.get_loc('signal')] = np.where(
            signals['short_mavg'].iloc[self.long_window:] > signals['long_mavg'].iloc[self.long_window:], 1.0, 0.0
        )

        # Generate trading orders using .diff()
        signals['positions'] = signals['signal'].diff()

        self.signals = signals
        print("Signal generation complete.")
        return self.signals[['positions']] # Return only the positions column for clarity

# --- 3. Order Management Component (Simulation) ---
class OrderManagement:
    """Simulates order placement and tracking."""
    def __init__(self):
        self.orders = {} # {order_id: {'symbol': ..., 'type': ..., 'side': ..., 'amount': ..., 'price': ..., 'status': ...}}
        self._next_order_id = 1

    def place_order(self, symbol, order_type, side, amount, price=None):
        """Simulates placing an order."""
        order_id = f"sim_order_{self._next_order_id}"
        self._next_order_id += 1
        status = 'filled' # Simulate instant fill for simplicity

        order_details = {
            'symbol': symbol,
            'type': order_type,
            'side': side,
            'amount': amount,
            'price': price if price is not None else self._simulate_fill_price(symbol, side),
            'status': status,
            'timestamp': pd.Timestamp.now()
        }
        self.orders[order_id] = order_details
        print(f"Simulated order placed and filled: {order_id} - {side} {amount:.4f} {symbol} at {order_details['price']:.2f}")
        return order_id, status, order_details

    def _simulate_fill_price(self, symbol, side):
        """Simulates a fill price based on current market conditions (placeholder)."""
        # In a real system, this would fetch the actual market price
        print("Simulating fill price...")
        # For simulation, let's use a random price around a plausible value
        # This is a very basic simulation; a real one would use actual market data
        return 60000 + (random.random() - 0.5) * 1000


    def cancel_order(self, order_id):
        """Simulates canceling an order."""
        if order_id in self.orders and self.orders[order_id]['status'] == 'open':
            self.orders[order_id]['status'] = 'canceled'
            print(f"Simulated order canceled: {order_id}")
            return True
        # print(f"Could not cancel order {order_id}: not found or not open.") # Comment out for cleaner test output
        return False

    def get_order_status(self, order_id):
        """Simulates getting order status."""
        return self.orders.get(order_id, {}).get('status', 'not_found')

# --- 4. Risk Management Component (Basic) ---
class RiskManagement:
    """Simulates basic risk management (e.g., tracking position)."""
    def __init__(self, initial_capital=100000):
        self.initial_capital = initial_capital
        self.current_capital = initial_capital
        # Use dictionaries for positions for better structure
        self.positions = {} # {symbol: {'buy': amount, 'sell': amount, 'avg_buy_price': price, 'avg_sell_price': price}}
        self.trade_history = []

    def update_position(self, trade_details):
        """Updates simulated position based on a filled trade."""
        symbol = trade_details['symbol']
        side = trade_details['side']
        amount = trade_details['amount']
        price = trade_details['price']
        timestamp = trade_details['timestamp']

        if symbol not in self.positions:
             self.positions[symbol] = {'buy': 0.0, 'sell': 0.0, 'avg_buy_price': 0.0, 'avg_sell_price': 0.0}

        if side == 'buy':
            total_amount = self.positions[symbol]['buy'] + amount
            # Calculate new average buy price
            if total_amount > 0:
                self.positions[symbol]['avg_buy_price'] = (
                    (self.positions[symbol]['buy'] * self.positions[symbol]['avg_buy_price']) + (amount * price)
                ) / total_amount
            self.positions[symbol]['buy'] = total_amount
            cost = amount * price
            self.current_capital -= cost
            print(f"Risk Management: Updated position - Bought {amount:.4f} of {symbol} at {price:.2f}. Current Capital: {self.current_capital:.2f}")
        elif side == 'sell':
            total_amount = self.positions[symbol]['sell'] + amount
             # Calculate new average sell price
            if total_amount > 0:
                 self.positions[symbol]['avg_sell_price'] = (
                     (self.positions[symbol]['sell'] * self.positions[symbol]['avg_sell_price']) + (amount * price)
                 ) / total_amount
            self.positions[symbol]['sell'] = total_amount
            revenue = amount * price
            self.current_capital += revenue
            print(f"Risk Management: Updated position - Sold {amount:.4f} of {symbol} at {price:.2f}. Current Capital: {self.current_capital:.2f}")

        self.trade_history.append(trade_details)

    def check_risk_rules(self):
        """Simulates checking basic risk rules (placeholder)."""
        # Implement stop-loss, max drawdown, etc. here
        # For now, just print current capital and positions
        # print(f"Risk Management: Current Capital = {self.current_capital:.2f}") # Comment out for cleaner output
        # print(f"Risk Management: Current Positions = {self.positions}") # Comment out for cleaner output
        pass # No risk rules enforced in this basic simulation

# --- 5. Integrated Trading System Simulation ---
class TradingSystem:
    """Integrates the components for simulation."""
    def __init__(self, symbol='BTC/USDT', initial_capital=100000):
        self.data_acquisition = DataAcquisition(symbol=symbol)
        self.strategy_execution = StrategyExecution()
        self.order_management = OrderManagement()
        self.risk_management = RiskManagement(initial_capital=initial_capital)
        self.symbol = symbol
        self.historical_data = None

    def run_backtest(self):
        """Runs a backtest simulation using historical data."""
        print("Running backtest simulation...")
        self.historical_data = self.data_acquisition.get_historical_data()

        if self.historical_data.empty:
            print("No historical data to backtest.")
            return

        # Ensure strategy has enough data
        min_data_points = max(self.strategy_execution.short_window, self.strategy_execution.long_window)
        if len(self.historical_data) < min_data_points:
            print(f"Not enough historical data for the strategy. Need at least {min_data_points} data points.")
            return

        signals = self.strategy_execution.generate_signals(self.historical_data)

        # Simulate trading based on signals
        print("Simulating trades based on signals...")
        # Reset risk management for backtest
        self.risk_management = RiskManagement(initial_capital=self.risk_management.initial_capital)

        # Start simulating from the first point where signals are valid
        start_index = max(self.strategy_execution.short_window, self.strategy_execution.long_window) -1

        for i in range(start_index, len(signals)):
            timestamp = signals.index[i]
            position_signal = signals['positions'].iloc[i]
            current_price = self.historical_data.loc[timestamp, 'price']

            if position_signal == 1.0: # Buy signal
                # Check if we are already in a buy position (simplified check)
                if self.risk_management.positions.get(self.symbol, {}).get('buy', 0) == 0:
                    # Buy with a fixed percentage of current capital (e.g., 10%)
                    amount_to_buy_usdt = self.risk_management.current_capital * 0.10
                    if current_price > 0: # Avoid division by zero
                        amount_to_buy = amount_to_buy_usdt / current_price
                    else:
                        amount_to_buy = 0

                    if amount_to_buy > 0:
                        # Simulate placing a market order
                        order_id, status, trade_details = self.order_management.place_order(
                            symbol=self.symbol,
                            order_type='market',
                            side='buy',
                            amount=amount_to_buy,
                            price=current_price # Use current price as fill price in backtest
                        )
                        if status == 'filled':
                            self.risk_management.update_position(trade_details)

            elif position_signal == -1.0: # Sell signal
                 # Check if we have a buy position to sell (simplified check)
                 amount_to_sell = self.risk_management.positions.get(self.symbol, {}).get('buy', 0)
                 if amount_to_sell > 0:
                    # Simulate placing a market order to sell the entire bought position
                    order_id, status, trade_details = self.order_management.place_order(
                        symbol=self.symbol,
                        order_type='market',
                        side='sell',
                        amount=amount_to_sell,
                        price=current_price # Use current price as fill price in backtest
                    )
                    if status == 'filled':
                        self.risk_management.update_position(trade_details)
                        # Reset buy position after selling
                        self.risk_management.positions[self.symbol]['buy'] = 0.0
                        # Optionally reset average buy price after selling entire position
                        # self.risk_management.positions[self.symbol]['avg_buy_price'] = 0.0


            # Check risk rules after each simulated trade (optional during backtest)
            # self.risk_management.check_risk_rules()

        print("\nBacktest simulation finished.")
        print(f"Final Capital: {self.risk_management.current_capital:.2f}")
        print(f"Total P/L: {self.risk_management.current_capital - self.risk_management.initial_capital:.2f}")
        print("Trade History:")
        if self.risk_management.trade_history:
            for trade in self.risk_management.trade_history:
                # Ensure timestamp is a Timestamp object before formatting
                timestamp_str = trade['timestamp'].strftime('%Y-%m-%d %H:%M:%S') if isinstance(trade['timestamp'], pd.Timestamp) else str(trade['timestamp'])
                print(f"- {timestamp_str} | {trade['side'].upper()} {trade['amount']:.4f} {trade['symbol']} at {trade['price']:.2f} (Order ID: {trade.get('id', 'N/A')})")
        else:
            print("No trades were executed during the backtest.")


# --- 6. Test Cases ---

print("--- Running Test Cases ---")

# Test Data Acquisition
print("\nTesting Data Acquisition...")
data_acq_test = DataAcquisition(start_date='2024-01-01', end_date='2024-01-10')
test_data = data_acq_test.get_historical_data()
print(f"Simulated data shape: {test_data.shape}")
print(f"Simulated data head:\n{test_data.head()}")
latest_price = data_acq_test.get_latest_price()
print(f"Latest simulated price: {latest_price:.2f}")
assert not test_data.empty, "Data Acquisition Test Failed: Data is empty."
assert 'price' in test_data.columns, "Data Acquisition Test Failed: 'price' column missing."
print("Data Acquisition Test Passed.")

# Test Strategy Execution
print("\nTesting Strategy Execution...")
# Use a larger date range for strategy test to ensure enough data points for SMAs
data_acq_strategy_test = DataAcquisition(start_date='2023-01-01', end_date='2024-01-01')
strategy_test_data = data_acq_strategy_test.get_historical_data()
strategy_test = StrategyExecution(short_window=50, long_window=200)
test_signals = strategy_test.generate_signals(strategy_test_data)
print(f"Generated signals shape: {test_signals.shape}")
print(f"Generated signals head:\n{test_signals.head()}")
print(f"Generated signals tail:\n{test_signals.tail()}")
assert not test_signals.empty, "Strategy Execution Test Failed: Signals are empty."
assert 'positions' in test_signals.columns, "Strategy Execution Test Failed: 'positions' column missing."
# Basic check: Are there any non-zero signals? (May fail if data doesn't produce signals)
# It's better to check if the signals DataFrame is created correctly
assert isinstance(test_signals, pd.DataFrame), "Strategy Execution Test Failed: Output is not a DataFrame."
print("Strategy Execution Test Passed.")

# Test Order Management
print("\nTesting Order Management...")
order_mgmt_test = OrderManagement()
order_id_buy, status_buy, trade_details_buy = order_mgmt_test.place_order('ETH/USDT', 'market', 'buy', 0.1)
# print(f"Order placed: {order_id_buy}, Status: {status_buy}") # Output handled by place_order
assert status_buy == 'filled', "Order Management Test Failed: Buy order not filled."
assert order_id_buy in order_mgmt_test.orders, "Order Management Test Failed: Buy order not recorded."

order_id_sell, status_sell, trade_details_sell = order_mgmt_test.place_order('ETH/USDT', 'market', 'sell', 0.05)
# print(f"Order placed: {order_id_sell}, Status: {status_sell}") # Output handled by place_order
assert status_sell == 'filled', "Order Management Test Failed: Sell order not filled."
assert order_id_sell in order_mgmt_test.orders, "Order Management Test Failed: Sell order not recorded."

# Test cancel (will fail as orders are simulated as instantly filled)
cancel_success = order_mgmt_test.cancel_order(order_id_buy)
# print(f"Cancel buy order success: {cancel_success}") # Output handled by cancel_order
assert cancel_success is False, "Order Management Test Failed: Cancel succeeded on filled order."

status_check = order_mgmt_test.get_order_status(order_id_sell)
print(f"Status of sell order {order_id_sell}: {status_check}")
assert status_check == 'filled', "Order Management Test Failed: Get status incorrect."

print("Order Management Test Passed (with simulated instant fills).")


# Test Risk Management
print("\nTesting Risk Management...")
risk_mgmt_test = RiskManagement(initial_capital=50000)
# Simulate a buy trade
sim_buy_trade = {
    'symbol': 'XRP/USDT', 'side': 'buy', 'amount': 100.0, 'price': 0.5,
    'timestamp': pd.Timestamp.now(), 'id': 'sim_trade_buy_1'
}
risk_mgmt_test.update_position(sim_buy_trade)
assert risk_mgmt_test.current_capital == 50000 - (100.0 * 0.5), "Risk Management Test Failed: Capital not reduced after buy."
assert 'XRP/USDT' in risk_mgmt_test.positions, "Risk Management Test Failed: Position not recorded."
assert risk_mgmt_test.positions['XRP/USDT']['buy'] == 100.0, "Risk Management Test Failed: Buy amount incorrect."
assert risk_mgmt_test.positions['XRP/USDT']['avg_buy_price'] == 0.5, "Risk Management Test Failed: Avg buy price incorrect."


# Simulate another buy trade for averaging test
sim_buy_trade_2 = {
    'symbol': 'XRP/USDT', 'side': 'buy', 'amount': 50.0, 'price': 0.6,
    'timestamp': pd.Timestamp.now(), 'id': 'sim_trade_buy_2'
}
risk_mgmt_test.update_position(sim_buy_trade_2)
expected_avg_buy_price = ((100.0 * 0.5) + (50.0 * 0.6)) / (100.0 + 50.0)
assert risk_mgmt_test.positions['XRP/USDT']['buy'] == 150.0, "Risk Management Test Failed: Total buy amount incorrect."
assert risk_mgmt_test.positions['XRP/USDT']['avg_buy_price'] == expected_avg_buy_price, "Risk Management Test Failed: Avg buy price calculation incorrect."


# Simulate a sell trade
sim_sell_trade = {
    'symbol': 'XRP/USDT', 'side': 'sell', 'amount': 50.0, 'price': 0.6,
    'timestamp': pd.Timestamp.now(), 'id': 'sim_trade_sell_1'
}
risk_mgmt_test.update_position(sim_sell_trade)
assert risk_mgmt_test.positions['XRP/USDT']['sell'] == 50.0, "Risk Management Test Failed: Sell amount incorrect."
assert len(risk_mgmt_test.trade_history) == 3, "Risk Management Test Failed: Trade history incorrect." # 1 initial buy + 1 second buy + 1 sell

risk_mgmt_test.check_risk_rules() # Just print check
print("Risk Management Test Passed.")

# Test Integrated System Simulation (Backtest)
print("\nTesting Integrated Trading System (Backtest)...")
# Use a larger date range for a more meaningful backtest
trading_system_test = TradingSystem(symbol='TEST/SIM', initial_capital=200000) # Removed start_date and end_date here as it's handled by DataAcquisition
trading_system_test.run_backtest()
# Basic assertion: Final capital should be non-negative
assert trading_system_test.risk_management.current_capital >= 0, "Integrated System Test Failed: Final capital is negative."
print("Integrated Trading System Test Completed (review output manually).")

print("\n--- All Test Cases Finished ---")
```

## Implement Risk Management and Monitoring

### Subtask:
Build in robust risk management protocols (e.g., stop-loss, position sizing, maximum drawdown) and continuous monitoring of system performance and market conditions.

```
# --- Extend Risk Management Component ---

class AdvancedRiskManagement:
    """Extends basic RiskManagement with more comprehensive rules and monitoring."""
    def __init__(self, initial_capital=100000, max_drawdown_percent=10, max_position_size_usdt=10000):
        self.initial_capital = initial_capital
        self.current_capital = initial_capital
        self.max_drawdown_percent = max_drawdown_percent
        self.max_position_size_usdt = max_position_size_usdt
        self.peak_capital = initial_capital
        self.positions = {} # {symbol: {'buy': amount, 'sell': amount, 'avg_buy_price': price, 'avg_sell_price': price}}
        self.trade_history = []
        self.is_system_active = True # Flag to halt trading if risk limits are breached

    def update_position(self, trade_details):
        """Updates simulated position and capital based on a filled trade."""
        if not self.is_system_active:
            print("Risk Management: System is halted due to risk breach. Not updating position.")
            return

        symbol = trade_details['symbol']
        side = trade_details['side']
        amount = trade_details['amount']
        price = trade_details['price']
        timestamp = trade_details['timestamp']

        if symbol not in self.positions:
             self.positions[symbol] = {'buy': 0.0, 'sell': 0.0, 'avg_buy_price': 0.0, 'avg_sell_price': 0.0}

        if side == 'buy':
            total_amount = self.positions[symbol]['buy'] + amount
            if total_amount > 0:
                self.positions[symbol]['avg_buy_price'] = (
                    (self.positions[symbol]['buy'] * self.positions[symbol]['avg_buy_price']) + (amount * price)
                ) / total_amount
            self.positions[symbol]['buy'] = total_amount
            cost = amount * price
            self.current_capital -= cost
            print(f"Risk Management: Updated position - Bought {amount:.4f} of {symbol} at {price:.2f}. Current Capital: {self.current_capital:.2f}")
        elif side == 'sell':
            total_amount = self.positions[symbol]['sell'] + amount
            if total_amount > 0:
                 self.positions[symbol]['avg_sell_price'] = (
                     (self.positions[symbol]['sell'] * self.positions[symbol]['avg_sell_price']) + (amount * price)
                 ) / total_amount
            self.positions[symbol]['sell'] = total_amount
            revenue = amount * price
            self.current_capital += revenue
            print(f"Risk Management: Updated position - Sold {amount:.4f} of {symbol} at {price:.2f}. Current Capital: {self.current_capital:.2f}")

        self.trade_history.append(trade_details)
        self._check_risk_rules() # Check rules after every trade

    def _check_risk_rules(self):
        """Checks risk management rules and halts the system if necessary."""
        # Check Maximum Drawdown
        self.peak_capital = max(self.peak_capital, self.current_capital)
        drawdown = (self.peak_capital - self.current_capital) / self.peak_capital * 100 if self.peak_capital > 0 else 0

        if drawdown > self.max_drawdown_percent:
            print(f"Risk Management ALERT: Maximum drawdown of {self.max_drawdown_percent}% breached!")
            print(f"Current Drawdown: {drawdown:.2f}%")
            self._halt_system("Maximum drawdown breached")

        # Check Max Position Size (Simplified - per symbol, buy side)
        for symbol, pos_data in self.positions.items():
            if pos_data['buy'] > 0 and pos_data['avg_buy_price'] is not None:
                 current_position_value = pos_data['buy'] * pos_data['avg_buy_price'] # Using average buy price as approximation
                 if current_position_value > self.max_position_size_usdt:
                     print(f"Risk Management ALERT: Maximum position size for {symbol} breached!")
                     print(f"Current Position Value ({symbol}): {current_position_value:.2f} USDT")
                     # Decide whether to halt or just log - halting is safer for this example
                     self._halt_system(f"Max position size for {symbol} breached")


    def _halt_system(self, reason="Risk management rule breached"):
        """Halts the trading system by setting a flag."""
        self.is_system_active = False
        print(f"Trading system halted: {reason}")
        # In a real system, you would also send alerts, close open positions, etc.

    def is_active(self):
        """Returns True if the trading system is active, False otherwise."""
        return self.is_system_active

# --- Basic Monitoring Functions (Conceptual) ---

def monitor_performance(risk_manager):
    """Simulates monitoring key performance indicators."""
    print("\n--- System Monitoring ---")
    print(f"Current Capital: {risk_manager.current_capital:.2f}")
    print(f"Initial Capital: {risk_manager.initial_capital:.2f}")
    print(f"Total P/L: {risk_manager.current_capital - risk_manager.initial_capital:.2f}")
    peak_capital = risk_manager.peak_capital
    drawdown = (peak_capital - risk_manager.current_capital) / peak_capital * 100 if peak_capital > 0 else 0
    print(f"Peak Capital: {peak_capital:.2f}")
    print(f"Current Drawdown: {drawdown:.2f}% (Max allowed: {risk_manager.max_drawdown_percent}%)")
    print(f"System Active: {risk_manager.is_active()}")
    print("--- End Monitoring ---")

# --- Example Usage with AdvancedRiskManagement ---

print("--- Testing Advanced Risk Management and Monitoring ---")

# Simulate trades with the advanced risk manager
advanced_risk_mgmt_test = AdvancedRiskManagement(initial_capital=100000, max_drawdown_percent=5, max_position_size_usdt=20000)

# Simulate a series of trades
sim_trade_1 = {'symbol': 'ABC/USDT', 'side': 'buy', 'amount': 10.0, 'price': 1000.0, 'timestamp': pd.Timestamp.now(), 'id': 'trade_1'} # Value 10,000
advanced_risk_mgmt_test.update_position(sim_trade_1)
monitor_performance(advanced_risk_mgmt_test)

sim_trade_2 = {'symbol': 'XYZ/USDT', 'side': 'buy', 'amount': 5.0, 'price': 3000.0, 'timestamp': pd.Timestamp.now(), 'id': 'trade_2'} # Value 15,000
advanced_risk_mgmt_test.update_position(sim_trade_2)
monitor_performance(advanced_risk_mgmt_test)

# Simulate a losing trade to trigger drawdown
sim_trade_3_loss = {'symbol': 'ABC/USDT', 'side': 'sell', 'amount': 5.0, 'price': 900.0, 'timestamp': pd.Timestamp.now(), 'id': 'trade_3'} # Simulate selling at a loss
advanced_risk_mgmt_test.update_position(sim_trade_3_loss)
monitor_performance(advanced_risk_mgmt_test)

# Simulate a trade that breaches max position size (if system is still active)
sim_trade_4_large = {'symbol': 'ABC/USDT', 'side': 'buy', 'amount': 15.0, 'price': 1000.0, 'timestamp': pd.Timestamp.now(), 'id': 'trade_4'} # Total ABC position value would be (10+15)*1000 = 25000
if advanced_risk_mgmt_test.is_active():
    advanced_risk_mgmt_test.update_position(sim_trade_4_large)
    monitor_performance(advanced_risk_mgmt_test)
else:
    print("\nAttempted to place a large trade, but system is halted.")


print("\n--- Advanced Risk Management and Monitoring Test Finished ---")
```
